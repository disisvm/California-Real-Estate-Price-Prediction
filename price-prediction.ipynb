Constant columns, which contain the same value across all rows, can skew analysis and are not useful for identifying outliers. Therefore, we identify and remove these:

## Outlier Removal in Numeric Columns

### Removing Outliers Using IQR

For each numeric column in the dataset, outliers are identified and removed using the Interquartile Range (IQR) method. The IQR is a measure of statistical dispersion and is calculated as the difference between the 75th percentile (Q3) and the 25th percentile (Q1) of the data:

This code calculates the lower and upper bounds for acceptable values in each column. Data points that fall outside of these bounds are considered outliers and are excluded from the dataset. This filtering is done within the same loop to ensure all specified numeric columns are cleaned.

### Displaying Cleaned Data Information

After outlier removal, the cleaned dataset's structure is displayed to provide a quick summary of the data.

The output includes details about the remaining entries and columns, showing how many entries have survived the outlier removal process. It lists the total number of non-null entries for each column, their data types, and memory usage, giving an overview of what the dataset looks like after cleaning.

````python
for column in numeric_columns:
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

# Display cleaned data information
cleaned_data_info = data.info()
cleaned_data_info
````

# Enhancing Feature Set with Interaction and Polynomial Features

### Creating Interaction Features

Interaction features are created to explore potential synergies between different variables, which might enhance the predictive power of a model.

These features combine basic attributes to form more complex ones, like the number of bathrooms per bedroom, or the living area per room, potentially highlighting aspects that simple features might miss.

### Adding Polynomial Features

Polynomial features allow models to capture more complex relationships between features by considering non-linear combinations of the features.

This transformation creates a new set of features which include the square of each feature and the interaction term between `livingArea` and `bedrooms`.

### Merging Polynomial Features with Original Data

The new polynomial features are concatenated with the original numeric data to enrich the dataset.

### Removing Duplicate Columns

To ensure that there are no duplicate columns which could skew the analysis, we keep only the first occurrence of each column.

### Displaying the Enhanced Dataset

To verify the enhancements, the first few rows of the updated dataset are displayed.

### Summary

By integrating interaction and polynomial features, the dataset gains additional dimensions that may reveal new insights and improve the performance of predictive models. This approach is especially useful in regression and classification contexts where relationships between variables can significantly impact the outcome.

````python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures

# Create interaction features
numeric_data['bed_bath_rooms'] = numeric_data['bedrooms'] * numeric_data['bathrooms']
numeric_data['living_area_per_room'] = numeric_data['livingArea'] / numeric_data['bedrooms']
numeric_data['living_area_per_bathroom'] = numeric_data['livingArea'] / numeric_data['bathrooms']
numeric_data['living_area_per_rooms'] = numeric_data['livingArea'] / (numeric_data['bedrooms'] + numeric_data['bathrooms'])
numeric_data['garage_spaces_per_bedroom'] = numeric_data['garageSpaces'] / numeric_data['bedrooms']

# Add polynomial features for livingArea and bedrooms
poly = PolynomialFeatures(degree=2, include_bias=False)
poly_features = poly.fit_transform(numeric_data[['livingArea', 'bedrooms']])
poly_features_df = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(['livingArea', 'bedrooms']))

# Concatenate the polynomial features with the original dataset
numeric_data = pd.concat([numeric_data, poly_features_df], axis=1)

# Remove duplicate columns by keeping the first occurrence
numeric_data = numeric_data.loc[:, ~numeric_data.columns.duplicated()]

# Display the first few rows of the new dataset
numeric_data.head()
````

# Machine Learning Pipeline Setup for Real Estate Price Prediction

### Importing Necessary Libraries

The code begins by importing various modules from `scikit-learn` for preprocessing, model building, and evaluation.

### Defining Feature Categories

Features are categorized into numeric and categorical types to apply appropriate preprocessing steps.

### Preprocessing Pipelines

Separate preprocessing pipelines are set up for numeric and categorical features:

- **Numeric Pipeline:** Applies standard scaling to normalize the numeric features, aiding in convergence during model training.
- **Categorical Pipeline:** Uses one-hot encoding to convert categorical features into a format suitable for machine learning algorithms.

### Combining Preprocessing Steps

A `ColumnTransformer` combines the numeric and categorical pipelines.

This structure allows different preprocessing steps to be applied to specified columns of the dataset.

### Feature Selection and Data Preparation

Irrelevant features are dropped, and the dataset is divided into features (`X`) and the target variable (`y`).

### Data Transformation

The preprocessing pipeline is applied to the features, transforming the dataset into a format suitable for training.

### Splitting the Dataset

The preprocessed data is split into training and testing sets to evaluate the model's performance on unseen data.

### Summary

This setup prepares the data through a robust preprocessing pipeline, ensuring that both numeric and categorical data are appropriately scaled and encoded. The transformed data is then ready for fitting various regression models, aiming to predict real estate prices effectively.

````python
from sklearn.preprocessing import OneHotEncoder, StandardScaler, PolynomialFeatures
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Define numeric and categorical features

numeric_features = ['livingArea^2', 'livingArea', 'bathrooms', 
                    'livingArea bedrooms', 'bed_bath_rooms', 'living_area_per_room', 
                    'garageSpaces', 'bedrooms', 'bedrooms^2', 'living_area_per_rooms', 'hasGarage', 
                    'longitude', 'living_area_per_bathroom', 'yearBuilt',
                    'latitude', 'garage_spaces_per_bedroom']

categorical_features = ['city', 'state', 'county', 'homeType']

# Create preprocessing pipelines for numeric and categorical data
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combine preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Define the features and target variable
features_to_drop = ['price', 'pricePerSquareFoot', 'livingAreaValue', 'id', 'datePostedString', 'description', 
                    'streetAddress', 'hasBadGeocode']
X = data.drop(columns=features_to_drop)
y = data['price']

# Apply the transformations
X_preprocessed = preprocessor.fit_transform(X)

# Split the preprocessed data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y, test_size=0.2, random_state=42)
````

## Model Training and Evaluation

### Linear Regression

The Linear Regression model is trained using the training dataset. Predictions are then made on the test set.

The model is evaluated based on Mean Squared Error (MSE) and R-squared values to assess its accuracy and the proportion of variance in the dependent variable that is predictable from the independent variables.





````python
# Train the Linear Regression model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Predict on the test set
y_pred_lr = lr_model.predict(X_test)

# Evaluate the model
mse_lr = mean_squared_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)

print(f'Linear Regression MSE: {mse_lr}')
print(f'Linear Regression R-squared: {r2_lr}')
````

### Decision Tree Regressor

Following a similar pattern, the Decision Tree Regressor is trained. Predictions are made on the test set.

````python
# Train the Decision Tree Regressor
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)

# Predict on the test set
y_pred_dt = dt_model.predict(X_test)

# Evaluate the model
mse_dt = mean_squared_error(y_test, y_pred_dt)
r2_dt = r2_score(y_test, y_pred_dt)

print(f'Decision Tree Regressor MSE: {mse_dt}')
print(f'Decision Tree Regressor R-squared: {r2_dt}')
````

### Random Forest Regressor

Lastly, the Random Forest Regressor is trained, which uses an ensemble of decision trees to improve the model’s predictive accuracy and control over-fitting.

### Summary

# Model Selection Based on Performance Metrics

### Setting Up Model Evaluation

A set of models, which includes the best-performing instances from the Random Forest, Gradient Boosting, and XGBoost regressors, is defined along with their respective MSE and R-squared values.

This setup facilitates an organized approach to comparing the models based on their performance metrics.

### Identifying the Best Model

Variables are initialized to keep track of the best model encountered during iteration:

- `best_mse` is set to infinity initially to ensure that any lower MSE will update this value.
- `best_r2` starts from 0 to make sure any model with a higher R-squared will improve upon this.

### Comparison Logic

A loop iterates over each model, updating the best model variables only if the current model has a lower MSE and a higher R-squared than previously recorded.

This conditional ensures that the selected best model not only fits the data closely (low MSE) but also explains a high proportion of variance in the data (high R-squared).

### Output the Best Model

Finally, the best model and its metrics are printed out.

### Summary

This approach systematically evaluates multiple regression models to select the one that optimally predicts the target variable with the least error and highest explanation of variance. It is an effective method for choosing the most suitable model for deployment in predictive tasks, such as real estate pricing, ensuring that decisions are made based on robust quantitative metrics.

````python
# Define the models and their metrics
models = [
    (best_rf_model, mse_rf, r2_rf),
    (best_gb_model, mse_gb, r2_gb),
    (best_xgb_model, mse_xgb, r2_xgb)
]

# Initialize variables to track the best model
best_model = None
best_mse = float('inf')
best_r2 = 0

# Iterate over the models to find the best one
for model, mse, r2 in models:
    if mse < best_mse and r2 > best_r2:
        best_model = model
        best_mse = mse
        best_r2 = r2

print(f'Best Model: {best_model}')
print(f'Best Model MSE: {best_mse}')
print(f'Best Model R-squared: {best_r2}')
````
